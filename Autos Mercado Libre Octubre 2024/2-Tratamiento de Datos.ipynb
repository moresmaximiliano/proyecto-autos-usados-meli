```python
import pandas as pd
import glob
import chardet

# Ruta a los archivos CSV 
file_pattern = 'autos/*.csv'

# Obtener una lista de todos los archivos CSV que coinciden con el patrón
csv_files = glob.glob(file_pattern)

# Crear una lista para almacenar los DataFrames
dataframes = []

# Función para detectar la codificación del archivo
def detect_encoding(file):
    with open(file, 'rb') as f:
        result = chardet.detect(f.read())
        return result['encoding']

# Leer cada archivo CSV con la codificación detectada
for file in csv_files:
    encoding = detect_encoding(file)
    df = pd.read_csv(file, encoding=encoding, sep=',')
    dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
df_autos = pd.concat(dataframes, ignore_index=True)

# df_autos.to_csv('autos.csv')

pd.set_option('display.max_columns', None)
df_autos
```

## Tratamos la columna Tipo de Combustible

```python
# Valores que quieres reemplazar
combustibles_a_reemplazar = ['Eléctrico', 'GNC', 'Híbrido', 'Híbrido/Diesel', 'Híbrido/Nafta', 'Nafta/GNC']

# Iterar sobre cada modelo
for modelo in df_autos['Modelo'].unique():
    # Filtrar por modelo
    mask_modelo = df_autos['Modelo'] == modelo
    
    # Calcular la moda de "Tipo de Combustible" para ese modelo (excluyendo los valores a reemplazar)
    moda_combustible = df_autos.loc[mask_modelo & ~df_autos['Tipo de combustible'].isin(combustibles_a_reemplazar), 'Tipo de combustible'].mode()
    
    # Si hay una moda, reemplazar los valores en la columna "Tipo de Combustible" para ese modelo
    if not moda_combustible.empty:
        df_autos.loc[mask_modelo & df_autos['Tipo de combustible'].isin(combustibles_a_reemplazar), 'Tipo de combustible'] = moda_combustible[0]
df_autos
```

## Tratamos la columna Transmisión

```python
# Cambiar los valores de la columna 'Transmisión' a Automático y Manual
df_autos.loc[df_autos['Transmisión'].isin(['Automática secuencial','AutomáticaMotor','Automático']), 'Transmisión'] = 'Automática'
df_autos.loc[df_autos['Transmisión'].isin(['ManualMotor','Semiautomática']), 'Transmisión'] = 'Manual'

# Rellenar los valores nulos de 'Transmisión' con la moda según el 'Modelo'
for modelo in df_autos['Modelo'].unique():
    # Filtrar los datos del modelo actual
    mask_modelo = df_autos['Modelo'] == modelo
    
    # Calcular la moda de 'Transmisión' para ese modelo (excluyendo valores nulos)
    moda_transmision = df_autos.loc[mask_modelo, 'Transmisión'].mode()
    
    # Si existe una moda, rellenar los valores nulos de 'Transmisión' para ese modelo
    if not moda_transmision.empty:
        df_autos.loc[mask_modelo & df_autos['Transmisión'].isna(), 'Transmisión'] = moda_transmision[0]

df_autos
```

## Tratamos la columna Motor 

```python
# Asignamos motores según la descripción de la versión
import re

# Función para extraer el primer número que sigue el patrón número.número en cualquier parte de la cadena
def extraer_motor(version, motor_actual):
    # Intentamos encontrar el patrón de número seguido de un punto y otro número en cualquier parte de la cadena
    match = re.search(r'(\d+\.\d+)', str(version))
    
    # Si encontramos un match, devolvemos el número extraído
    if match:
        return match.group(0)
    else:
        # Si no hay un número en la versión, devolvemos el motor que ya estaba
        return motor_actual

# Aplicar la función a cada fila del DataFrame utilizando .loc
df_autos.loc[:, 'Motor'] = df_autos.apply(lambda row: extraer_motor(row['Versión'], row['Motor']), axis=1)

# Rellenar los valores nulos de 'Motor' con la moda según el 'Modelo'
for modelo in df_autos['Modelo'].unique():
    # Filtrar los datos del modelo actual
    mask_modelo = df_autos['Modelo'] == modelo
    
    # Calcular la moda de 'Motor' para ese modelo (excluyendo valores nulos)
    moda_motor = df_autos.loc[mask_modelo, 'Motor'].mode()
    
    # Si existe una moda, rellenar los valores nulos de 'Motor' para ese modelo
    if not moda_motor.empty:
        df_autos.loc[mask_modelo & df_autos['Motor'].isna(), 'Motor'] = moda_motor[0]

# Asegurarse de que todos los valores de la columna 'Motor' sean numéricos usando .loc[]
df_autos.loc[:, 'Motor'] = pd.to_numeric(df_autos['Motor'], errors='coerce')

# Definir los valores válidos para la columna Motor
valores_validos = {1.0, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0, 2.2, 2.3, 2.4, 2.5, 2.8, 3.0, 3.2, 3.6, 3.8, 4.0, 4.4}

# Filtrar el DataFrame para mantener solo los valores válidos en la columna Motor
df_autos = df_autos[df_autos['Motor'].isin(valores_validos)]

df_autos
```

## Tratamos la columna Tipo de Carrocería

```python
# Asignamos carrocería según el modelo
# Función para asignar la moda del tipo de carrocería según el modelo
def asignar_moda_carroceria(row, df):
    # Filtrar el DataFrame para el mismo modelo
    mask_modelo = df['Modelo'] == row['Modelo']
    
    # Calcular la moda de "Tipo de carrocería" para ese modelo
    moda_carroceria = df.loc[mask_modelo, 'Tipo de carrocería'].mode()
    
    # Si existe una moda, devolverla; si no, devolver el valor original
    if not moda_carroceria.empty:
        return moda_carroceria[0]
    return row['Tipo de carrocería']

# Aplicar la función utilizando .loc para asignar los valores sin la advertencia
df_autos.loc[:, 'Tipo de carrocería'] = df_autos.apply(lambda row: asignar_moda_carroceria(row, df_autos), axis=1)

df_autos
```

## Tratamos la columna Capacidad del tanque

```python
# Extraer el valor numérico de la columna 'Capacidad del tanque'
df_autos['Capacidad del tanque'] = df_autos['Capacidad del tanque'].str.extract(r'(\d+)').astype(float)

# Rellenar los valores nulos con la moda según el modelo
df_autos['Capacidad del tanque'] = df_autos.groupby('Modelo')['Capacidad del tanque'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x))

df_autos
```

## Tratamos la columna Potencia

```python
# Rellenar los valores nulos en 'Potencia' con la moda según el 'Modelo'
df_autos['Potencia'] = df_autos.groupby('Modelo')['Potencia'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x))

df_autos
```

## Tratamos las columnas Largo, Altura, Ancho

```python
# Rellenar los valores nulos en 'Largo', 'Ancho', 'Altura' con la moda según 'Modelo'
columnas_a_rellenar = ['Largo', 'Ancho', 'Altura']

for columna in columnas_a_rellenar:
    df_autos[columna] = df_autos.groupby('Modelo')[columna].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x))

df_autos
```

## Tratamos la columna Pilóto Automático y Aire Acondicionado

```python
# Función para rellenar con la moda según el modelo, y 'No' si no hay moda
def rellenar_con_moda_o_no(df, columna):
    return df.groupby('Modelo')[columna].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'No'))

# Rellenar los valores nulos en 'Piloto automático' y 'Aire acondicionado'
df_autos['Piloto automático'] = rellenar_con_moda_o_no(df_autos, 'Piloto automático')
df_autos['Aire acondicionado'] = rellenar_con_moda_o_no(df_autos, 'Aire acondicionado')

df_autos
```

## Tratamos la columna Distancia entre Ejes

```python
# Eliminar el texto ' mm' y convertir a valor numérico
df_autos['Distancia entre ejes'] = df_autos['Distancia entre ejes'].str.replace(' mm', '').astype(float)

# Función para rellenar con la moda según el modelo, y 'No' si no hay moda
def rellenar_con_moda(df, columna):
    return df.groupby('Modelo')[columna].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mean()))

# Rellenar los valores nulos en 'Distancia entre ejes' con la moda según el modelo
df_autos['Distancia entre ejes'] = rellenar_con_moda(df_autos, 'Distancia entre ejes')

# Asignar el valor 2729.0 en 'Distancia entre ejes' para los modelos 'Clase GLA'
df_autos.loc[df_autos['Modelo'] == 'Clase GLA', 'Distancia entre ejes'] = 2729.0

df_autos
```

## Tratamos la columna Año

```python
# Eliminar los valores de 'Año' que sean iguales a 1906 o mayores a 2024
df_autos = df_autos[(df_autos['Año'] != 1906) & (df_autos['Año'] <= 2024)]

df_autos
```

## Tratamos la columna Precio

```python
# Precios desde 1900 hasta 190000 dejar como están (USD), y de 2000000 hasta 70000000 dividir por 1300 (DOLAR BLUE) (eliminar 111111)
# Eliminar las filas con precio igual a 111111
df_autos = df_autos[df_autos['Precio'] != 111111]

# Filtrar los precios que estén entre 1900 y 190000, o entre 2000000 y 70000000
df_autos = df_autos[((df_autos['Precio'] >= 1900) & (df_autos['Precio'] <= 190000)) | 
                    ((df_autos['Precio'] >= 2000000) & (df_autos['Precio'] <= 70000000))]

# Dividir los precios entre 2000000 y 70000000 por 1300
df_autos['Precio'] = df_autos['Precio'].apply(lambda x: x / 1300 if x >= 2000000 else x)

df_autos
```

```python
# Lista de las columnas categóricas para las cuales se crearon variables dummy
columnas_categoricas = ['Ubicación', 'Marca', 'Modelo', 'Tipo de combustible', 'Transmisión', 
                         'Tipo de carrocería', 'Dirección', 'Control de tracción', 
                         'Piloto automático', 'Aire acondicionado']

# Crear las variables dummy sin eliminar la primera columna (drop_first=False)
df_autos = pd.get_dummies(df_autos, 
                          columns=columnas_categoricas,
                          drop_first=False)

# Convertimos todas las columnas booleanas a enteros
df_autos = df_autos.map(lambda x: 1 if x is True else (0 if x is False else x))

# Ahora, por cada columna original, eliminamos la dummy con menos valores 1
for col in columnas_categoricas:
    # Extraemos las columnas dummy relacionadas con la columna original
    dummies = [column for column in df_autos.columns if column.startswith(col + '_')]
    
    # Contamos la cantidad de 1s en cada dummy y encontramos la que menos 1s tiene
    dummy_counts = df_autos[dummies].sum()
    columna_a_eliminar = dummy_counts.idxmin()
    
    # Eliminamos la columna con menos valores 1
    df_autos = df_autos.drop(columns=[columna_a_eliminar])

# Colocamos la columna 'Precio' al final, si no lo estaba
precio_columna = df_autos.pop('Precio')
df_autos['Precio'] = precio_columna

df_autos
```
