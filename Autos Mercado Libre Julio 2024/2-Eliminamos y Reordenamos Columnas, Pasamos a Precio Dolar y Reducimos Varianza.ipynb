```python
# Tratamos de unir todos los csv en un solo df
import pandas as pd
import glob

# Ruta a los archivos CSV 
file_pattern = 'Autos/*.csv'  # Pongo para que me agregue todos los archivos csv de la carpeta

# Obtener una lista de todos los archivos CSV que coinciden con el patrón
csv_files = glob.glob(file_pattern)

# Crear una lista para almacenar los DataFrames
dataframes = []

# Leer cada archivo CSV y almacenarlo en la lista de DataFrames
for file in csv_files:
    df = pd.read_csv(file, encoding='utf-16', sep=';') # si tenemos strings, para convertir en listas agg converters={'Título':literal_eval, 'Ubicación':literal_eval,etc}
    dataframes.append(df)                              # usar converters si tenemos frases o palabras que se nos muestran como string en el df ('')

# Concatenar todos los DataFrames en uno solo
df_autos = pd.concat(dataframes, ignore_index=True)

df_autos
```

```python
# Eliminamos las columnas de las que no recolectamos datos
df_autos = df_autos.drop(columns=['Único dueño', 'Piloto automático', 'Aire acondicionado'])
df_autos # tenemos 41936 datos, de 40 modelos en 18 columna
```

```python
# Ordenamos las columnas según nuestro criterio
column_order = [
    'Título', 'Ubicación', 'Versión', 'Color', 'Marca', 'Modelo', 'Año',
    'Tipo de combustible', 'Puertas', 'Transmisión', 'Motor', 'Tipo de carrocería',
    'Kilómetros', 'Dirección', 'Control de tracción', 'Capacidad de personas',
    'Potencia', 'Precio'
]

# Reorganizar las columnas del DataFrame según el orden especificado
df_autos = df_autos[column_order]

df_autos
```

```python
# Definir una función para eliminar outliers según el 95% de concentración de precios por modelo
def eliminar_outliers(df, col_modelo, col_precio):
    # Crear un DataFrame vacío para almacenar los datos filtrados
    df_filtrado = pd.DataFrame()

    # Iterar sobre cada grupo de modelo
    for modelo, grupo in df.groupby(col_modelo):
        # Calcular los percentiles 2.5% y 97.5%
        lower_bound = grupo[col_precio].quantile(0.025)
        upper_bound = grupo[col_precio].quantile(0.975)
        
        # Filtrar los datos para el modelo actual
        grupo_filtrado = grupo[(grupo[col_precio] >= lower_bound) & (grupo[col_precio] <= upper_bound)]
        
        # Agregar los datos filtrados al DataFrame final
        df_filtrado = pd.concat([df_filtrado, grupo_filtrado], ignore_index=True)
    
    return df_filtrado

# Aplicar la función a tu DataFrame
df_autos = eliminar_outliers(df_autos, 'Modelo', 'Precio')

# Ver los datos filtrados
df_autos
```

```python
# Con esto convertimos los precios en pesos a precios en dólares
# Guardamos los nuevos datos del modelo con precios pasados a dólares 
# Filtrar los autos con modelo (DF_AUTOS)
autos = df_autos[df_autos['Modelo'] == 'C4'].copy() # cambiar según modelo

# Eliminar las filas con precios específicos
precios_a_eliminar = [140000,2200] # precios que comprobamos que no son de vehículos
autos = autos[~autos['Precio'].isin(precios_a_eliminar)]

# Dividir los precios mayores a 1000000 por 1500
autos.loc[autos['Precio'] > 1000000, 'Precio'] = autos.loc[autos['Precio'] > 1000000, 'Precio'] / 1500 # tipo de cambio blue (CAMBIAR RANGO)
#autos.loc[autos['Precio'] >= 400000, 'Precio'] = autos.loc[autos['Precio'] >= 400000, 'Precio'] / 1500 # tipo de cambio blue (CAMBIAR RANGO)

# Redondear los precios a dos decimales
autos['Precio'] = autos['Precio'].round(2)

# Mostrar los datos filtrados con precios redondeados
#autos

# Guardar el DataFrame filtrado y redondeado en un archivo CSV
autos.to_csv('Autos a Dólar/autos_c4.csv', index=False) # CAMBIAR SEGÚN MODELO
```
